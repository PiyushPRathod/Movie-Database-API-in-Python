Rendering the movie image

```python
import json
import requests
from pprint import pprint

api_key = 'ee7ce7a118cf1e19e66ad23499d268f5'

response = requests.get("https://api.themoviedb.org/3/movie/238/images",
                        params={'api_key':api_key}).json()

#Storing details only one backdrop for our example
backdrop = response['backdrops'][0]

#Extract the file path
file_path = backdrop['file_path']

#Base image URL
base_image_url = "https://image.tmdb.org/t/p/"
#The size of image
img_size = "w500"

final_image_url = base_image_url + img_size + file_path

print(f"<img src={final_image_url} style='margin-left:200px; margin-top:20px'>")
```

Rendering the video using YouTube

```python
import json
import requests
from pprint import pprint

api_key = 'ee7ce7a118cf1e19e66ad23499d268f5'

response = requests.get("https://api.themoviedb.org/3/movie/238/videos",
                        params={'api_key':api_key}).json()

#Extract the key
key = response['results'][0]['key']

#YouTubue's embed URL
embed_url = "https://www.youtube.com/embed/"

final_video_url = embed_url + key

print(f"<div class='embed-responsive embed-responsive-16by9'> <iframe class='embed-responsive-item' style='height: 400px; width:800px; margin-left:50px' src={final_video_url}></iframe> </div>")
```
